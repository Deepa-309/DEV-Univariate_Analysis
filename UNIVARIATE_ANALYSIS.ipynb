Plotting a Lorenz Curve

```python
import numpy as np

# Sort income data
sorted_income = np.sort(df_income['Income'])
cumulative_income = np.cumsum(sorted_income) / np.sum(sorted_income)
cumulative_population = np.linspace(0, 1, len(sorted_income))

# Plot Lorenz curve
plt.figure(figsize=(8,6))
plt.plot(cumulative_population, cumulative_income, label='Lorenz Curve')
plt.plot([0, 1], [0, 1], linestyle='--', label='Line of Equality')
plt.title('Lorenz Curve for Income Distribution')
plt.xlabel('Cumulative Population')
plt.ylabel('Cumulative Income')
plt.legend()
plt.show()
```

The Gini Coefficient

```python
# Simulate Gini calculation
def gini_coefficient(data):
    sorted_data = np.sort(data)
    cumulative_data = np.cumsum(sorted_data) / np.sum(sorted_data)
    return 1 - 2 * np.trapezoid(cumulative_data, dx=1/len(sorted_data))

gini_income = gini_coefficient(df_income['Income'])
print(f"Gini Coefficient for Income: {gini_income:.2f}")
```

Alternative Measures of Inequality

```python
from scipy.stats import entropy

# Calculate Theil index (entropy-based)
def theil_index(data):
    return entropy(data) / len(data)

theil_income = theil_index(df_income['Income'])
print(f"Theil Index for Income: {theil_income:.2f}")
```

Smoothing

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Simulated temperature data
data = {
    'Date': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04', '2021-01-05'],
    'Temperature': [32, 35, 28, 30, 33]  # Ensure this is named 'Temperature'
}

# Create DataFrame
df = pd.DataFrame(data)

# Ensure 'Date' is in the correct format
df['Date'] = pd.to_datetime(df['Date'])

# Set 'Date' as index
df.set_index('Date', inplace=True)

# Check if 'Temperature' column exists
print(df.columns)

# Ensure the 'Temperature' column exists and apply Exponential Smoothing
if 'Temperature' in df.columns:
    model = ExponentialSmoothing(df['Temperature'], trend=None, seasonal=None)
    fitted_model = model.fit()

    # Add the smoothed values to the DataFrame
    df['Smoothed'] = fitted_model.fittedvalues

    # Plot original and smoothed series
    plt.figure(figsize=(10, 6))
    plt.plot(df.index, df['Temperature'], label='Original')
    plt.plot(df.index, df['Smoothed'], label='Exponentially Smoothed', color='red')
    plt.xlabel('Date')
    plt.ylabel('Temperature')
    plt.title('Exponential Smoothing')
    plt.legend()
    plt.show()
else:
    print("Column 'Temperature' not found in the DataFrame.")
```

 Lowess Smoothing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Example data for df_poll
data = {
    'Month': ['2021-01', '2021-02', '2021-03', '2021-04', '2021-05',
              '2021-06', '2021-07', '2021-08', '2021-09', '2021-10'],
    'Approval Rating': [45, 47, 50, 49, 51, 53, 52, 54, 56, 55]
}

# Create a DataFrame
df_poll = pd.DataFrame(data)

# Convert 'Month' column to datetime
df_poll['Month'] = pd.to_datetime(df_poll['Month'])

# Apply LOWESS smoothing
lowess = sm.nonparametric.lowess(df_poll['Approval Rating'], np.arange(len(df_poll)))

# Plot original data and LOWESS smoothed trend
plt.figure(figsize=(10, 6))
plt.plot(df_poll['Month'], df_poll['Approval Rating'], label='Original Data')
plt.plot(df_poll['Month'], lowess[:, 1], label='LOWESS Smoothed', color='red')
plt.xlabel('Month')
plt.ylabel('Approval Rating')
plt.title('LOWESS Smoothing of Opinion Poll')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout for better fit
plt.show()
```

Residuals in Smoothing Time Series

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Simulate time series data
np.random.seed(0)
data = {'Month': pd.date_range(start='2023-01-01', periods=12, freq='M'),
        'Sales': np.random.poisson(200, 12) + np.linspace(0, 50, 12)}
df = pd.DataFrame(data)
df.set_index('Month', inplace=True)

# Apply a moving average
df['Moving Average'] = df['Sales'].rolling(window=3).mean()

# Calculate residuals
df['Residuals'] = df['Sales'] - df['Moving Average']

# Plot
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Sales'], label='Actual Sales', marker='o')
plt.plot(df.index, df['Moving Average'], label='Moving Average', color='orange')
plt.plot(df.index, df['Residuals'], label='Residuals', color='red', linestyle='--')
plt.title('Sales, Moving Average, and Residuals')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

Refinements in Smoothing Time Series

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.api import ExponentialSmoothing

# Extended data for Sales over time to ensure at least two full seasonal cycles
data = {
    'Month': [
        '2021-01', '2021-02', '2021-03', '2021-04', '2021-05',
        '2021-06', '2021-07', '2021-08', '2021-09', '2021-10',
        '2021-11', '2021-12', '2022-01', '2022-02', '2022-03',
        '2022-04', '2022-05', '2022-06', '2022-07', '2022-08',
        '2022-09', '2022-10', '2022-11', '2022-12', '2023-01'
    ],
    'Sales': [
        200, 210, 215, 225, 230, 235, 220, 240, 245, 250,
        260, 270, 275, 280, 290, 295, 300, 310, 320, 330,
        340, 350, 360, 370, 380
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Convert 'Month' column to datetime and set as index
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)

# Fit an Exponential Smoothing model with a seasonal component
model = ExponentialSmoothing(df['Sales'], seasonal='add', seasonal_periods=12)
fit_model = model.fit()

# Predict and calculate residuals
df['Exp Smooth'] = fit_model.fittedvalues
df['Exp Residuals'] = df['Sales'] - df['Exp Smooth']

# Plot
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Sales'], label='Actual Sales', marker='o')
plt.plot(df.index, df['Exp Smooth'], label='Exponential Smoothing', color='orange')
plt.plot(df.index, df['Exp Residuals'], label='Exponential Residuals', color='red', linestyle='--')
plt.title('Sales, Exponential Smoothing, and Residuals')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to fit elements better
plt.show()
```
